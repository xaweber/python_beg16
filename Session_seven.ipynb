```python
import pandas as pd
```

```python
surveys_df = pd.read_csv('surveys.csv')
```

```python
species_df = pd.read_csv("species.csv")
```

```python
survey_sub = surveys_df.head(10)
survey_sub
```

```python
survey_sub_last = surveys_df[-11:-1]
survey_sub_last
```

```python
survey_sub_last = survey_sub_last.reset_index(drop=True)
survey_sub_last
```

```python
vertical_stack = pd.concat([survey_sub, survey_sub_last], axis = 0)
vertical_stack
```

```python
vertical_stack = pd.concat([survey_sub, survey_sub_last], axis = 1)
vertical_stack
```

```python
vertical_stack.to_csv('out.csv')
```

```python
survey_80 = surveys_df[surveys_df.year == 1980]
survey_82 = surveys_df[surveys_df.year == 1982]
```

```python
survey_8082 = pd.concat([survey_80, survey_82])
survey_8082 = survey_8082.reset_index(drop=True)
survey_8082.head()
```

```python
survey_8082.to_csv('survey_1980_1982.csv')
```
